```python
#pip install citipy
#Import Modules and keys
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from citipy import citipy
from config import owm_key
import random
```

```python
seedCount = int(input(f"How many random numbers ya need? "))
cityLatSeeds= random.sample(range(-10000,10000),seedCount)
cityLngSeeds= random.sample(range(-10000,10000),seedCount)
x=0.0
cityLat = [(x/10000)*90 for x in cityLatSeeds]
cityLng = [(x/10000)*180 for x in cityLngSeeds]

cityList=[]

for x in range(0,seedCount):
    cityList.append(citipy.nearest_city(cityLat[x],cityLng[x]).city_name)

cities_df = pd.DataFrame({
    'City':cityList,
    'Latitude':cityLat,
    'Longitude':cityLng
})
cities_df['Dupes'] = cities_df.duplicated(subset='City')
cities_df
cities_clean=cities_df.loc[cities_df['Dupes']==False]
cities_clean=cities_clean[['City','Latitude','Longitude']]
cities_clean
```
