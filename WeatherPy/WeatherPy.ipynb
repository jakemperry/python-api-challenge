```python
#pip install citipy
#Import Modules and keys
import os
import requests
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from citipy import citipy
from config import owm_key
import random
import json
```

```python
#randos = int(input(f"How many random numbers ya need? "))
randos=2000
cityLatSeeds= random.sample(range(-10000,10000),randos)
cityLngSeeds= random.sample(range(-10000,10000),randos)
x=0.0
cityLat = [(x/10000)*90 for x in cityLatSeeds]
cityLng = [(x/10000)*180 for x in cityLngSeeds]

cityList=[]

for x in range(0,randos):
    cityList.append(citipy.nearest_city(cityLat[x],cityLng[x]).city_name)

cities_df = pd.DataFrame({
    'City':cityList,
    'Latitude':cityLat,
    'Longitude':cityLng
})
cities_df['Dupes'] = cities_df.duplicated(subset='City')
cities_df
cities_clean=cities_df.loc[cities_df['Dupes']==False]
cities_clean.reset_index()
cities_clean=cities_clean[['City','Latitude','Longitude']]
cities_clean
```

```python
cityNum = int(input(f"How many cities do you want? "))
allCitiesList = list(citipy.WORLD_CITIES_DICT.values())

allCitiesNames = [x.city_name for x in allCitiesList]
cityList3 = random.sample(allCitiesNames,cityNum)

# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={owm_key}&units={units}&q="

# set up lists to hold reponse info
lat = []
lng = []
temp = []
humidity = []
cloudiness = []
windspeed=[]
cityFound=[]
i=1
# Loop through the list of cities and perform a request for data on each
for city in cityList3:
    try:
        print(f"Requesting city {i} of {cityNum}: {city}")
        response = requests.get(query_url + city).json()
        cityFound.append(response['name'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        temp.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        windspeed.append(response['wind']['speed'])
        
    except: 
        print(f'Couldn\'t get all the data for {city}.  On to the next city...')
        cityList3
    i+=1
weather_dict = {
    "City": cityFound,
    "Lat": lat,
    'Lng':lng,
    "Temp_F": temp,
    'Humidity':humidity,
    'Cloudiness':cloudiness,
    'Windspeed':windspeed
}
weather_data = pd.DataFrame(weather_dict)
weather_data.to_csv(os.path.join('temp',f'WeatherData_{len(weather_data)}.csv'))
weather_data.head()
```
