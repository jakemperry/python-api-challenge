```python
#pip install citipy
#Import Modules and keys
import os
import requests
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from citipy import citipy
from config import owm_key
from scipy import stats
import random
import json
```

```python
cityNum = int(input(f"How many cities do you want? "))
allCitiesList = list(citipy.WORLD_CITIES_DICT.values())

allCitiesNames = [x.city_name for x in allCitiesList]
cityListDraft = random.sample(allCitiesNames,cityNum)
cityListFinal=list(dict.fromkeys(cityListDraft))
if len(cityListFinal) <500:
    print(f"Less than 500 cities.  Enter a larger number for the initial city set")
else:
    print(f"A list of {len(cityListFinal)} cities is ready to go!")
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={owm_key}&units={units}&q="

# set up lists to hold reponse info
lat = []
lng = []
temp = []
humidity = []
cloudiness = []
windspeed=[]
cityFound=[]
i=1
# Loop through the list of cities and perform a request for data on each
for city in cityListFinal:
    try:
        print(f"Requesting city {i} of {cityNum}: {city}")
        response = requests.get(query_url + city).json()
        cityFound.append(response['name'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        temp.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        windspeed.append(response['wind']['speed'])
        
    except: 
        print(f'Couldn\'t get all the data for {city}.  On to the next city...')
    i+=1
weather_dict = {
    "City": cityFound,
    "Lat": lat,
    'Lng':lng,
    "Temp_F": temp,
    'Humidity':humidity,
    'Cloudiness':cloudiness,
    'Windspeed':windspeed
}
randomWeather = pd.DataFrame(weather_dict)
randomWeather.to_csv(os.path.join('temp',f'WeatherData_{len(randomWeather)}.csv'))
randomWeather.head()
```

```python
#Run this cell for a copy of a saved CSV file from a previous set of requests.
#This way you'll generate the same figures as the ones from the initial set of calls (750 requests, 715 responses)
randomWeather=pd.read_csv(os.path.join('temp','WeatherData_703_2020_0205.csv'))
randomWeather=randomWeather[['City','Lat','Lng','Temp_F','Humidity','Cloudiness','Windspeed']]
randomWeather.head()
```

```python
randomWeather.plot(kind='scatter',y='Temp_F',x='Lat')
plt.grid(b=True,which='major',axis='both')
plt.ylabel('Temperature (F)')
plt.xlabel('Latitude')
plt.xlim(-90,90)
plt.title('Temperature vs Latitude (5 Feb 2021)')
plt.tight_layout()
plt.show()
print(f"Temperature appears to increase closer to the equator.  Temperature also appears to be \ncooler in the Northern Hemisphere as it is currently winter in the Northern Hemisphere \n(February 2021).  There appear to be more data points in the Northern Hemisphere, \nbut this may be because there is more land mass in the Northern Hemisphere, so there are \nmore places to put weather stations than in the Southern Hemisphere.")
```

```python
randomWeather.plot(kind='scatter',y='Humidity',x='Lat')
plt.grid(b=True,which='major',axis='both')
plt.ylabel('Humidity (%)')
plt.xlabel('Latitude')
plt.xlim(-90,90)
plt.title('Humidity vs Latitude (5 Feb 2021)')
plt.tight_layout()
plt.show()
print(f"Again, there are more data points in the Northern Hemisphere compared to the \nSouthern Hemisphere, likely due to more land mass up North.  Humidity appears \nto be lower near the equator, which may be due to dryer areas like the \nsouthwest United States and Saharan Africa.")

```

```python
randomWeather.plot(kind='scatter',y='Cloudiness', x='Lat')
plt.grid(b=True,which='major',axis='both')
plt.ylabel('Cloudiness (% Cloud Cover)')
plt.xlabel('Latitude')
plt.xlim(-90,90)
plt.title('Cloudiness vs Latitude (5 Feb 2021)')
plt.tight_layout()
plt.show()
print(f"Cloud coverage appears to vary at all latitudes.  It also appears there may \nbe some variations in data sources, as many of the cloudiness reports seem \nto fall along 0%, 20%, 40%, 75%, 90%, and 100%.  Perhaps there are different \nstandards for measuring the percentage of cloud cover.  The cloud coverage \ndoes not appear to have a strong correlation to latitude.")
```

```python
randomWeather.plot(kind='scatter',y='Windspeed',x='Lat')
plt.grid(b=True,which='major',axis='both')
plt.ylabel('Wind speed (mph)')
plt.xlabel('Latitude')
plt.xlim(-90,90)
plt.title('Wind Speed vs Latitude (5 Feb 2021)')
plt.tight_layout()
plt.show()
print(f"Wind speed appears to be somewhat higher in the Northern Hemisphere than \nin the Southern Hemisphere, and appears to have lower values at the equator.  \nAs this data was pulled during February 2021, it's possible that the wind \nspeed is higher in the Northern Hemisphere because it's currently winter in \nthe Northern Hemisphere.")
```

```python
#Make two new dataframes, one for northern hemisphere(Lat>=0) and southern hemisphere (Lat<0)
weatherNH=randomWeather.loc[randomWeather['Lat']>=0]
weatherSH=randomWeather.loc[randomWeather['Lat']<0]
```

```python
weatherNH.plot(kind='scatter',y='Temp_F',x='Lat')
plt.grid(b=True,which='major',axis='both')
plt.ylabel('Temperature (F)')
plt.xlabel('Latitude')
plt.xlim(0,90)
plt.title('Temperature vs Latitude, Northern Hemisphere (5 Feb 2021)')
slope, y_int, r, p, std_err = stats.linregress(weatherNH['Lat'], weatherNH['Temp_F'])
fit = slope * weatherNH['Lat'] + y_int
line_eq=f'y={round(slope,2)}x+{round(y_int,2)}, \nr= {round(r,2)}'
plt.annotate(line_eq,(65,80),fontsize=10,color="red",xytext=(65,80))
plt.plot(weatherNH['Lat'],fit,"-",color='red')
plt.tight_layout()
plt.show()
```
